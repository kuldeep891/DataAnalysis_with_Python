```python
import pandas as pd
```

## Create 3 Datasets

```python
dummy_data1 = {
        'id': ['1', '2', '3', '4', '5'],
        'Feature1': ['A', 'C', 'E', 'G', 'I'],
        'Feature2': ['B', 'D', 'F', 'H', 'J']}
```

```python
df1 = pd.DataFrame(dummy_data1, columns = ['id','Feature1','Feature2'])
df1.head(2)
```

```python
dummy_data2 = {
        'id': ['1', '2', '6', '7', '8'],
        'Feature1': ['K', 'M', 'O', 'Q', 'S'],
        'Feature2': ['L', 'N', 'P', 'R', 'T']}
```

```python
df2 = pd.DataFrame(dummy_data2, columns = ['id','Feature1','Feature2'])
df2.head(2)
```

```python
dummy_data3 = {
        'id': ['1', '2', '3', '4', '5', '7', '8', '9', '10', '11'],
        'Feature3': [12, 13, 14, 15, 16, 17, 15, 12, 13, 23]}
```

```python
df3 = pd.DataFrame(dummy_data3, columns = ['id','Feature3'])
df3.head(2)
```

## Concatenate df1 and df2
Rownumbering is incorrect once you concat, to rectify the row numbers we need  to use ignore_index = True

```python
#df_row = pd.concat([df1,df2])
df_row = pd.concat([df1,df2],ignore_index = True)
df_row
```

## Concatenate using Keys 
Keys separate the dataframes apart

```python
df_row = pd.concat([df1,df2], keys=['df1','df2'])
df_row
```

```python
df_row.loc['df1']
```

## Join two DataFrames

```python
df_merge = pd.merge(df_row,df3,how='inner',on='id')
df_merge
```

--> WHen the dataFrames have different names of the columns on which join is needed to apply
--> Use left_on and right_on parameters

```python
df_merge = pd.merge(df_row,df3,how='inner',left_on='id',right_on='id')
df_merge
```

```python
df_merge = pd.merge(df1,df2,how='inner',left_on='id',right_on='id',suffixes=['_left','_right'])
df_merge
```

## Add a row to the DataFrame
Row can be appended to the DataFrame by passing the row as Series or Dictionary

```python
add_row = pd.Series([9,'K','L',20],index=['id','Feature1','Feature2','Feature3'])
df_add_row=df_merge.append(add_row,ignore_index = True)
df_add_row
```
